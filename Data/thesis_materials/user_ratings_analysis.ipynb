## Each action, slow piece

In general, users were easily able to keep a steady tempo, which produced a realistic experience. Users were also generally able to speed up the orchestra as desired, although the response was sometimes unrealistic. On the other hand, slowing down the orchestra was both difficult and unrealistic for users, and was met with plenty of negative feedback. Finally, users had mixed experiences with accents.


```python
# Slow piece
# Order of samples:
# Joey Ishika Sebastian | Elysia
# Ishika didn't answer ribbons for time constraints

data = {
    'Easy to gesture':
    {
        'Speed up': [6, 5, 5, 5.5],
        'Slow down': [2, 5, 2.5, 2],
        'Steady': [7, 6, 5, 3],
        'Accent': [3, 5, 3.5, 4],
    },
    'Realistic response':
    {
        'Speed up': [5, 4, 4],
        'Slow down': [4, 4, 2.5],
        'Steady': [6, 6, 5] ,
        'Accent': [3, 6, 4],
    },

}

means = dict()
errors=dict()
for system in data:
    system_means = []
    system_errors = []
    for attr_vals in data[system].values():
        system_means.append(np.mean(attr_vals))
        system_errors.append(stats.sem(attr_vals))
    means[system] = system_means
    errors[system] = system_errors
```

```python
categories = ("Speed up", "Slow down", "Steady", "Accent")

x = np.arange(len(categories))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0.5

# fig, ax = plt.subplots(layout='constrained')
fig, ax = plt.subplots()
for attribute, measurement in means.items():
    print(attribute, measurement)
    offset = width * multiplier
    if ENABLE_ERROR_BARS:   
        rects = ax.bar(x + offset, measurement, width, label=attribute, yerr=errors[attribute], ecolor='grey')
        ax.bar_label(rects, padding=1, fmt = '%.1f', label_type='center')
    else:
        rects = ax.bar(x + offset, measurement, width, label=attribute)
        ax.bar_label(rects, padding=1, fmt = '%.1f')
    multiplier += 1
ax.set_ylabel('Rating')
ax.set_title('Ratings of Actions, Slow Piece\n(Higher value indicates more favorable rating)')
ax.set_xticks(x + width, categories)
ax.legend(loc='upper left', ncols=1)
ax.set_ylim(1, 7)
# ax.set_xlim(-0.5, 2)

plt.show()
```

Users were generally able to keep a steady tempo as desired, and the orchestra responded as expected too. Users noted that while trying to stay at a steady tempo, they especially received strong conducting practice.

Users also could speed up as desired. However, the response was criticized as not particularly realistic. In particular, users reported that the orchestra sometimes would speed up by drastic amounts beyond what was intended. Further, the orchestra could be made to play extremely fast, beyond human capabilities.

Slowing down proved to be a pain point to users. Users generally initially struggled to slow down the orchestra, and took the longer to get comfortable with this action than any others. Users especially noted that after many gestures of conducting a slower beat than the orchestra was playing, they could be suddenly conducting a beat off. This made the system hard to use, and was unrealistic to human playing, as well.

Users had moderate reviews of accenting beats. Even after adjusting parameters, users still could not always have accents played on the beats they intended for. Other times, accents would be played on beats when not intended, which was not only hard to use but also especially unrealistic. Users did enjoy the realism of increasing the MIDI notes' velocity, and of accents only occuring on beats where they might be expected by a real orchestra.

Hand vs Controller Tracking

```python
# Note that data is printed as inverse of recorded for this one only
# Effectively captured data on 1-7 when I should've on 7-1
# To transform, subtract from 8 for symmetric flip
data_easier = [8-1, 8-4, 8-5, 8-2.5]
mu_easier =  np.mean(data_easier)
se_easier = stats.sem(data_easier)

data_realistic = [8-1, 8-2, 8-2.5]
mu_realistic = np.mean(data_realistic)
se_realistic = stats.sem(data_realistic)
```

```python

categories = ("Easier gestures", "More realistic")
means = {
    'Rating': (mu_easier, mu_realistic)
}

x = np.arange(len(categories))  # the label locations
width = 0.25  # the width of the bars
multiplier = 1

# fig, ax = plt.subplots(layout='constrained')
fig, ax = plt.subplots()
for attribute, measurement in means.items():
    print(attribute, measurement)
    offset = width * multiplier
    if ENABLE_ERROR_BARS:   
        rects = ax.bar(x + offset, measurement, width, label=attribute, yerr=errors[attribute], ecolor='grey')
        ax.bar_label(rects, padding=1, fmt = '%.1f', label_type='center')
    else:
        rects = ax.bar(x + offset, measurement, width, label=attribute)
        ax.bar_label(rects, padding=1, fmt = '%.1f')
    multiplier += 1
ax.set_ylabel('Rating')
ax.set_title('Hand vs Controller Usage in VRC\n(Higher value favors hand usage)')
ax.set_xticks(x + width, categories)
# ax.legend(loc='upper left', ncols=3)
ax.set_ylim(1, 7)
ax.set_xlim(-0.5, 2)

plt.show()
```

On both counts, users find hand tracking to provide a better system than controller tracking. 

Gesture tracking and detection is slightly stronger with hand-tracking. Users enjoy the lighter weight of a baton than a controller. With the baton, moving the object and user's hand requires less force, making gestures easier to motion. However, this was slightly countered by hand-tracking's low framerate, high latency, and occasional bugs. 

Users especially note that holding a baton is very different from holding a controller, requiring different grips, wrist and hand movements, and spatial awareness. This makes controller-based conducting quite unrealistic to the real-life experience, and less helpful in training as well.

In general, VRC rates fairly well; its ratings are both competitive with its compeition and fairly high numerically. VRC is generally  rated slightly below BatonX, but not by much. Both systems are followed by RibbonsVR.

While all three systems do not fare greatly in realism, VRC is slightly outrated by BatonX. Users are not completely satisfied by BatonX's visuals, as time-stretching a prerecorded video feed could be painful to watch. Still, users find that this still positively enhances realism, and is especially better than the lack of any orchestral video feedback of VRC.

VRC does significantly outperform other systems in usability. Users point out relative strengths of VRC's accurate gesture detection, and its interface for navigating playback control and changing settings.

VRC is rated well in its ability as a training tool, with high scores both in helpfulness and recommendation potential to beginning conductors. Unfortunately, VRC falls slightly behind BatonX in both categories. According to users, BatonX's video walkthrough of conducting instructions, from within the system itself, are greatly helpful for teaching.
